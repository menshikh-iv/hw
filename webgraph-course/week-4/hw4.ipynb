```python
import networkx as nx

g = nx.read_edgelist("4.graph.txt", nodetype=int)
```

```python
import itertools as it


def disjoined(u, v, w):
    if g.has_edge(u, v): 
        return False

    if g.has_edge(v, w):
        return False

    if g.has_edge(w, u):
        return False
    
    return True
    
    
def nei_intersection(u, v, w):
    return set(g.neighbors(u)) & set(g.neighbors(v)) & set(g.neighbors(w))


with open("triplets.txt", 'w') as outfile:
    for idx, (u, v, w) in enumerate(it.combinations(g.nodes(), 3)):
        if (idx + 1) % 1000000 == 0:
            print(idx + 1)
        
        if not disjoined(u, v, w):
            continue
            
        comm_nei = nei_intersection(u, v, w)
        if len(comm_nei) < 3:
            continue

        outfile.write("{},{},{}\n".format(u, v, w))
```

```python
with open("triplets.txt") as infile:
    triplets = [map(int, line.strip().split(",")) for line in infile]
```

```python
def disjoined(u, v, w):
    if g.has_edge(u, v): 
        return False

    if g.has_edge(v, w):
        return False

    if g.has_edge(w, u):
        return False
    
    return True
    
```

```python
import itertools as it

def tri_nei_int(t1, t2):
    u0, v0, w0 = t1
    u1, v1, w1 = t2
    
    if g.has_edge(u0, u1) and g.has_edge(u0, v1) and g.has_edge(u0, w1) and\
       g.has_edge(v0, u1) and g.has_edge(v0, v1) and g.has_edge(v0, w1) and\
       g.has_edge(w0, u1) and g.has_edge(w0, v1) and g.has_edge(w0, w1) and\
       len(set(t1) | set(t2)) == 6 and\
       disjoined(u0, v0, w0) and disjoined(u1, v1, w1):
            return True
    
    
      
            
    return False
```

```python
result = []

for idx, t1 in enumerate(triplets):
    for idy, t2 in enumerate(triplets):
        if idx >= idy:
            continue
            
        if len(set(t1) & set(t2)):
            continue
            
        if tri_nei_int(t1, t2):
            result.append((t1, t2))
            
len(result)
```
